```python
import cv2

# 1. Load the pre-trained Haar Cascade XML classifier

face_cascade = cv2.CascadeClassifier(r'haarcascade_frontalface_default.xml')

print(face_cascade.empty())
video_cap= cv2.VideoCapture(0)         # For video capture "0" for webcam and "1" for external sources (pen-drive, CD, DvD, etc).

# Check if the webcam opened successfully
if not video_cap.isOpened():
    print("Error: Could not open the video source.")
    exit()

while True:

    success, frame = video_cap.read()
    
    if not success:
        print("Failed to grab a frame.")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.1, 5)

    #print("Faces found:", len(faces))
    
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("Face Detection", frame)
    
    # Monitor the keyboard. Break the loop if the 'q' key is pressed.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        face = frame.copy()
        print("Quitting...")
        break

# Deallocate hardware resources and close active GUI windows
video_cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from deepface import DeepFace

def age_detection(face):

    result = DeepFace.analyze(face, actions=['age'], enforce_detection=False)

    age = result[0]['age']
    #print("Predicted Age:", age)

    if age >= 18:
        return "ALLOW"

    return "NOT ALLOW"

def face_detect (face):
    gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier(r'haarcascade_frontalface_default.xml')  
    face_coord = face_cascade.detectMultiScale(gray, 1.1, 5)
    
    print(face_cascade.empty())
    #print("Faces found:", len(faces))
    
    # Face Marking
    for (x, y, w, h) in face_coord:

        cv2.rectangle(face, (x, y), (x + w, y + h), (0, 255, 0), 2)

        result = age_detection(face[y:y+h, x:x+w])
        label = f"Decision: {result}"
        
        cv2.putText(face, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2, cv2.LINE_AA)
        
    return face

image = cv2.imread(r"C:\Users\abhat\Pictures\Camera Roll\WIN_20260614_19_56_52_Pro.jpg")
cv2.imshow("Face Detection", face_detect(image))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
